```python
df1=pd.read_csv('Category_005_data (1).csv')
df1.tail(75)
```

```python
df1['Order_Demand']=df1['Order_Demand'].str.replace('(',"")
df1['Order_Demand']=df1['Order_Demand'].str.replace(')',"")
```

```python
df1.index = pd.to_datetime(df1.index)
df1.index = df1.index.astype('datetime64[ns]')
```

```python
df1.head(5)
```

```python
df1['Order_Demand'] = df1['Order_Demand'].apply(lambda x: sum(map(int, x.split())))
df1 = df1.groupby(['Date'], as_index=False)['Order_Demand'].sum()
```

```python
df1.tail(20)
```

```python
df1 = df1.set_index('Date')
df1.index
```

```python
df1.index = pd.to_datetime(df1.index)
```

```python
df1.index
```

```python
y = df1['Order_Demand'].resample('MS').mean()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
y.plot(figsize=(12,5))
plt.xlabel('Date')
plt.ylabel('Orders')
plt.show()
```

```python
from pylab import rcParams
import statsmodels.api as sm
if y.isnull().sum() > 0:
    y = y.fillna(method='ffill')
rcParams['figure.figsize'] = 18, 8
decomposition = sm.tsa.seasonal_decompose(y, model='additive')
fig = decomposition.plot()
plt.show()
```

```python
df1.tail(10)
```

```python
import itertools
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
print('SARIMAX1: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX2: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX3: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX4: {} x {}'.format(pdq[2], seasonal_pdq[4]))

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            print('SARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(20, 10))
plt.show()
```

```python

pred = results.get_prediction(start=pd.to_datetime('2015-01-01'), dynamic=False)
pred_ci = pred.conf_int()

ax = y['2014':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='blue', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Order_Demand')
plt.legend()
plt.show()
```

```python

observed_monthly_sum = y.resample('M').sum()
predicted_monthly_sum = pred.predicted_mean.resample('M').sum()

monthly_comparison_df = pd.concat([observed_monthly_sum, predicted_monthly_sum], axis=1)
monthly_comparison_df.columns = ['Observed', 'Predicted']

monthly_comparison_df.to_csv('monthly_comparison_data.csv')
```

```python
y_truth = y['2015-01-01':]
y_forecasted = pred.predicted_mean
y_truth_12_months = y_truth[:12]
y_forecasted_12_months = y_forecasted[:12]

mse_12_months = ((y_forecasted_12_months - y_truth_12_months) ** 2).mean()
rmse_12_months = np.sqrt(mse_12_months)

print('RMSE for the first 12 months:', round(rmse_12_months, 2))
```

```python

y_truth_12_months = y_truth[:12]
observed_sum_12_months = y_truth_12_months.sum()
average_monthly_observed_12_months = observed_sum_12_months / 12

print('Sum of observed values for the first 12 months:', observed_sum_12_months)
print('Average monthly value for the first 12 months:', average_monthly_observed_12_months)
```

```python
percentage_error = (rmse_12_months / average_monthly_observed_12_months) * 100
accuracy = 100 - percentage_error

print('Percentage Error:', round(percentage_error, 2), '%')
print('Accuracy:', round(accuracy, 2), '%')
```

```python
pred_uc = results.get_forecast(steps=75)
pred_ci = pred_uc.conf_int()
ax = y.plot(label='observed', figsize=(16, 8))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Order_Demand')
plt.legend()
plt.show()
```
